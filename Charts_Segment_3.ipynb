```python
# Import the dataset 
df = pd.read_csv("cardio_complete.csv")
# remove the columns : Unnamed:0, id
df = df.drop(columns=["Unnamed: 0", "id"], axis= 1)
```

```python
df
```

```python
df = df.rename(columns = {"age" : "age(years)"})
df["age(years)"]= (df["age(years)"]/365).astype(int)
# updated_df=updated_df.round({'age(years)': 2})
# updated_df
df
```

```python
df
```

```python
# Binning age(years)

df["age_bins"]= pd.cut(df["age(years)"].values, bins=4, labels=False)
df
```

```python
df["age_bins"].describe()

```

```python
# value counts



df = df[df["age(years)"]> 30]
df

```

```python
age_counts = df["age(years)"].value_counts()

age_counts_df =pd.DataFrame(age_counts) 

age_counts_df

```

```python
age_counts_df= age_counts_df.rename({"age(years)" : "occurance_count"})
age_counts_df
```

```python
# groupby()

groupby_df = df.groupby(["age(years)"]).first()
groupby_df
```

```python
# merging
merged_df = groupby_df.merge(age_counts_df, left_index=True, right_index=True)
merged_df
```

```python
merged_df = merged_df.rename(columns={"age(years)" : "occurences"})
```

```python
# Rename index

merged_df.index.name = "Age(years)"

# testing

merged_df.head(5)
```

```python
merged_df["occurences"].describe()
```

```python
occurences_series = merged_df.occurences
```

```python
# creating bins for CVD occurences


cardio_bins = [0, 2000, 3000, 4000]

merged_df["CVD_bins"] = pd.cut(merged_df.occurences, cardio_bins)
```

```python

# Check the DF
merged_df.count()
```

```python
merged_df.occurences.groupby(pd.cut(merged_df.occurences, cardio_bins)).count()
```

```python
# testing sns with merged_df
plt.figure(figsize= (10,10))
sns.scatterplot(x= merged_df.index, y= "occurences", data = merged_df, size = "CVD_bins",
                hue= "CVD_bins", sizes= (50,200), palette= ["#17A589", "#9B59B6", "#D1010E"])
plt.title("Cardiovascular Disease Cases by Age", fontsize=20)
plt.xlabel("Age in years", fontsize= 16)
plt.ylabel("Number of cases of CVD", fontsize=16);
```
